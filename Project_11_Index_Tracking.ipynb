## Index Tracking - an Introduction

Plan: Track the DJI with only __10 Stocks__ (Tracking Portfolio).

Criteria to select Tracking Portfolio: Those 10 Stocks with the __lowest individual Tracking Error__ (TE) in the Training/Fitting Period.

- Training/Fitting Period: 2017 & 2018
- (Forward) Testing Period: 2019

Once we have identified the 10 Tracking Stocks we need to find their __optimal weights__ in the Tracking Portfolio that __minimize TE__.

```python
#why there is a need to track an index
#investors cannot directly invest into an index but they can invest in a fund for instance
#ETF that tracks the index
#vast majority of Dow jones ETF is just buy all 30 stocks with the respective faiths 
#that is not tracking at all, it just replicating the index
#for dow jones index it can be done in a very efficient way because only 30 large stocks that are very liquid and the transaction and trading costs are pretty low
#there are other indexes with thousands of stocks and many of them are not liquid at all
#buying all stocks would be too costly
#definitely more efficient to track those indexes as closely as possible with small subset
```

```python
#selected 10 stocks with the lowest individual Tracking error
#identified portfolio that best tracks the index in a training period and fitted the parameters of portfolio to the data in the training period
#does not guarantee that the tracking portfolio does a good jjob on new data
#need to further test tracking portfolio's ability to track the index and therefore there is testing period
#typically periods should be long enough to have sufficient data and as short as possible because portfolio's ability to attract the index can change over time 
```

## Selecting the Tracking Stocks

```python
#try to minimze the tracking error
#identify 10 stocks with lowest tracking error
#put steps of calculating mean and standard deviation of return differences and annualizing then with user defined function
```

```python
def tracking_ann(returns):
    summary = returns.agg(["mean", "std"]).T
    summary["Premium"] = summary["mean"] * 252
    summary["TE"] = summary["std"] * np.sqrt(252)
    summary.drop(columns = ["mean", "std"], inplace = True)
    return summary

```

```python
#calculate mean and std on daily basis 
#annualize average differences - annualized differences return(premium)
#premium can be either negative or positive
#annualize the standard deviation - tracking error
```

```python
tracking = tracking_ann(ret_diff)
tracking
```

## Index Tracking - A simple Tracking Portfolio

```python
tracking_stocks
```

```python
#created a simple tracking portfolio (equal weighted portfolio with 10 stocks)
#equal weights (starting point of optimization)
#each and every stock has weight of 10% 
#create numpy array with 10 time 10% with np.full
#shape is actually 10
#each and every value is 1/10 which is 10%
eweights = np.full(noa, 1/noa)
eweights
```

```python
#select only tracking stocks and daily returns and calculate simple average
#by passing axis=1, calcualting mean for each and every row 
ret2.loc[:, tracking_stocks].mean(axis = 1)
```

```python
#calculate portfolio returns for portfolios that are not equally weighted - use dot method
#starting with daily returns of 10 constituents
#calculate for each and every day, the weighted average return
ret2.loc[:, tracking_stocks].dot(eweights)
```

```python
#calculate portfolio returns (based on weights)
#pass weights array that sums up to 1
#function would return daily portfolio returns
def port_ret(weights):
    return ret2.loc[:, tracking_stocks].dot(weights)
```

```python
#define function to be minimized -> minimize tracking error of tracking portfolio
#pass weights array and the function returns annualized tracking error
#first, calculates portfolio returns and then subtract the return of dow jones index
#get return differences between tracking portfolio and dow jones index
#calculate standard deviation
#get tracking error on daily basis
#annualize tracking error 
def tracking_error(weights): 
     return port_ret(weights).sub(ret2.DJI).std() * np.sqrt(252)
```

```python
tracking_error(eweights)
#returned tracking error of 4.1%
```

```python
#compare with individual tracking errors of stocks
#calculate cumulative investment multiple for each and every time stamp-> calculate normalized prices by adding one and calculate cumulative product
#want to have base value or starting value of 100
#starting with the second day of the year
ew_port = port_ret(eweights)[1:].add(1).cumprod().mul(100)
ew_port
```

```python
#first trading day in 2017 is 3rd of january
```

```python
#copy ew_port to dataframe normalized tracking
#create a new column EW_Port
norm_track["EW_Port"] = ew_port
#make sure on the very first day starting at base value of 100
norm_track.iloc[0, -1] = 100
```

```python
norm_track
```

```python
norm_track[["EW_Port", "DJI"]].plot(figsize = (15, 10))
plt.legend(fontsize = 20, loc = 2)
plt.show()
```

## Index Tracking - The optimal Tracking Portfolio

```python
#try to find the optimal weights for 10 stocks
#use quadratic optimization algorithm
```

```python
import scipy.optimize as sco #import scipy optimize
```

```python
#constraint: weights must sum up to 1 -> sum of weights - 1 = 0
cons = ({"type": "eq", "fun": lambda x: np.sum(x) - 1})
cons
```

```python
#bounds: all weights shall be between 0 and 1 -> can be changed
#do not allow short selling
bnds =  tuple((0,1) for x in range(noa))
bnds
```

```python
#run optimization based on function to be minimized, starting with equal weights and based on respective bounds and constraints
#minimize tracking error
#define starting point 
#strating point of optimization actually equal weights
#SLSQP-quadratic optimization
opts = sco.minimize(tracking_error, eweights, method = "SLSQP", bounds = bnds, constraints= cons)
```

```python
#output of optimization
opts
#tracking error of optimal portfolio - 3.7%
#better than tracking error of equally weighted portfolio
```

```python
#getting the optimal weights
optimal_weights = opts["x"]
optimal_weights
```

```python
pd.Series(index = tracking_stocks, data = optimal_weights) #attach ticker symbols
```

```python
#daily returns of optimal portfolio by parsing optimal weights to function port_ret
#normalized values starting with base value of 100
opt_port = port_ret(optimal_weights)[1:].add(1).cumprod().mul(100)
opt_port
```

```python
norm_track["Opt_Port"] = opt_port
norm_track.iloc[0, -1] = 100
```

```python
norm_track
```

```python
#visualize optimal portfolio that actually best tracks the dow jones index
norm_track[["Opt_Port", "DJI"]].plot(figsize = (15, 10))
plt.legend(fontsize = 20, loc = 2)
plt.show()
```

```python
#compare optimal portfolio, equally weighted portfolio and dow jones index
#calculate annualized risk and return
summary_ann(norm_track[["Opt_Port", "EW_Port", "DJI"]].pct_change())
```

## Forward Testing (Part 1)

```python
df
```

```python
ret #dataframe returns
```

```python
tracking_stocks
```

```python
#automate and scale the process of calculating tracking errors
#visualizing tracking portfolios and index
#parameters - data frame with returns
#list of tracking stocks
#list or array with weights
#define index to be tracked
#define period - start and end date
#calculates and returns annualized tracking error between tracking stocks and weights and index to be tracked in the period from start to end 

def tracking_error_gen(data, tracking_stocks, weights, index, start, end):
    return data.loc[start:end, tracking_stocks].dot(weights).sub(ret.loc[start:end, index]).std() * np.sqrt(252)   
```

```python
#parameters - return data frame, tracking stocks list, list or array with weights, index to be tracked and period from start to end
#annualized tracking error in the training period
def visual(data, tracking_stocks, weights, index, start, end):
    track = data.loc[start:end, tracking_stocks].dot(weights).add(1).cumprod().mul(100)
    track.name = "Portfolio"
    index = data.loc[start:end, index].add(1).cumprod().mul(100)
    plt.figure(figsize = (12,8))
    track.plot()
    index.plot()
    plt.legend(fontsize = 20, loc = 2)
    plt.show()
```

```python

tracking_error_gen(ret, tracking_stocks, optimal_weights, "DJI", "2017-01-01", "2018-12-31")
#annualized tracking error of 3.79%
```

```python
#tracking error for year 2019
#forward test portfolio's ability to attract the index
forward_te = tracking_error_gen(ret, tracking_stocks, optimal_weights, "DJI", 
                                "2019-01-01", "2019-12-31")
forward_te
#forward tracking error is higher than back testing error - 4.7%
```

```python
visual(ret, tracking_stocks, optimal_weights, "DJI", "2019-01-01", "2019-12-31")
```

```python
#both charts closely move together even if the tracking error is slightly higher
#tracking portfolio works in the future as well
```

## Forward Testing (Part 2)

```python
#compare tracking portfolio with random portfolio
#if portfolio tracks the index better than the vast majority of random portfolios
#strongly indicates method to create tracking portfolio is valid and works with new data as well
#random portfolio - select 10 random stocks and generate random weights for stocks
bucket = df.columns.drop(["DOW", "DJI"])
bucket
#bucket to draw randomly 10 stocks
#bucket contains total 29 stocks
```

```python
np.random.choice(bucket, size = noa, replace = False)
#draw random samples
#replace = false
#random sampling without replacement (not to have stock twice in a sample)
```

```python
#10 random weights for the stocks that sum up to 1
#create 10 random floats
#dividing each and every random float by total sum - weights sum up to 1
random_numbers = np.random.random(noa)
random_weights = random_numbers / random_numbers.sum()
random_weights
#random_weights.sum()
```

```python
#setting a random seed guarantees reproducibility of random process
#create random stocks, random weights
np.random.seed(123)
random_stocks = np.random.choice(bucket, size = noa, replace = False)
random_numbers = np.random.random(noa)
random_weights = random_numbers / random_numbers.sum()
```

```python
#create 10000 random portfolios
#calculate 10000 tracking errors
#create empty numpy array to store 10000 tracking errors
#following this procedure for 10000 times
##create random weights and random stocks  - random portfolio
##calculate tracking error of random portfolio
#created 10000 random portfolios and in numpy array there will be 10000 tracking errors stored
tes = np.empty(10000)
np.random.seed(123)
for i in range(10000):
    random_numbers = np.random.random(noa)
    random_weights = random_numbers / random_numbers.sum()
    random_stocks = np.random.choice(bucket, size = noa, replace = False)
       
    tes[i] = tracking_error_gen(ret, random_stocks, random_weights, "DJI", "2019-01-01", "2019-12-31")
```

```python
(tes < forward_te).mean()
#how many random portfolios have tracking error lower than tracking portfolio
#for each and every random portfolio, check whether tracking error is lower than the forward tracking error of tracking portfolio 
##take them into calculated percentage

#6.5% of random portfolis are better than tracking portfolio
#strong indication
##tracking portfolio has some superior  abilities to track the index and it's not just random
```

```python
#histogram with distribution of 10000 tracking errors 
plt.figure(figsize = (12, 8))
plt.hist(tes, bins = 100)
plt.show()
```
